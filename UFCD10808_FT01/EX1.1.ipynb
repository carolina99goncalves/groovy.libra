# Revisão dos fundamentos de Python

## 1- Manipulação de Listas

### Lista "alturas"

```python
alturas = [154, 158, 165, 169, 176, 178, 161, 180, 167, 158, 169, 178, 176]
```

a) definir uma nova lista "tamanho_s", para as alturas entre 150 e 160

```python
tamanho_s = [i for i in alturas if 150 <= i <= 160]
print(tamanho_s)
```

b) Calcular o maior e menor valores para ambas as listas

```python
print("Maior altura:", max(alturas))
print("Menor altura:", min(alturas))
print("Maior tamanho_s:", max(tamanho_s))
print("Menor tamanho_s:", min(tamanho_s))
```

c) Calcular a soma de todos os elementos da lista "alturas"

```python
soma_alturas = sum(alturas)
print(soma_alturas)
```

d) Calcular a média da lista "alturas"

```python
media = sum(alturas) / len(alturas)
print("Média:", media)
```

e) Implementar a função "calcular_valores_unicos" que retorna a quantidade de elementos únicos na lista "alturas"

```python
def calcular_valores_unicos(lista):
    return len(set(lista))
print("Quantidade de valores únicos:", calcular_valores_unicos(alturas))
```

f) Adicionar dois valores aleatórios à lista "alturas"

```python
import random
alturas.extend([random.randint(150, 190) for _ in range(2)])
print(alturas)
```

g) Eliminar elementos com valor superior a 170

```python
alturas = [i for i in alturas if i < 170]
print(alturas)
```

h) Adicionar à lista "alturas" os valores [176, 178, 181]

```python
alturas.extend([176, 178, 181])
print(alturas)
```

i) Determinar a altura com mais ocorrências

```python
from collections import Counter
mais_comum = Counter(alturas).most_common(1)[0][0]
print("Altura mais frequente:", mais_comum)
```

j) Ordenar a lista "alturas"

```python
alturas.sort()
print(alturas)
```

k) Adicionar o elemento "altura" à lista "alturas"

```python
alturas.append("altura")
print(alturas)
```

l) Imprimir todos os elementos da lista "alturas"

```python
for altura in alturas:
    print(altura)
```
